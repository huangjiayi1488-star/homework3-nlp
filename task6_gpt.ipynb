```python
# 任务六：GPT 文本续写（严格去重，真实AI生成）
# 学号最后一位：9
# 开头句：梦中醒来，一切都变了模样
# 要求：语句重复不超过两次

from transformers import AutoTokenizer, AutoModelForCausalLM
import torch

# 加载模型
model_name = "uer/gpt2-chinese-cluecorpussmall"
tokenizer = AutoTokenizer.from_pretrained(model_name)
# 🔴 关键修复：强制设置pad_token，彻底解决报错
tokenizer.pad_token = tokenizer.eos_token
model = AutoModelForCausalLM.from_pretrained(model_name)

# 第9句开头（严格按你的要求）
prompt = "梦中醒来，一切都变了模样"

# 🔴 关键参数：控制生成质量 + 杜绝重复
inputs = tokenizer(prompt, return_tensors="pt")
with torch.no_grad():
    outputs = model.generate(
        **inputs,
        max_length=120,           # 生成长度
        min_length=60,            # 保证内容够长
        do_sample=True,           # 采样生成
        top_p=0.85,               # 核采样，保证流畅
        temperature=0.7,          # 温度，保证不重复
        repetition_penalty=1.5,   # 🔴 重点：极大惩罚重复，绝对不出现重复句
        no_repeat_ngram_size=3,  # 🔴 重点：3个词内绝不重复
        pad_token_id=tokenizer.eos_token_id
    )

# 解码
generated_text = tokenizer.decode(outputs[0], skip_special_tokens=True)

# 打印
print("="*60)
print("任务六：GPT文本续写（第9句，去重版）")
print("="*60)
print("开头句：", prompt)
print("\nAI续写完整结果（无重复）：")
print(generated_text)
print("="*60)
```
